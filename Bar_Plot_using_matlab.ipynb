```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
#Bar PLOT
#Bar plot that show how much  Spend time on programming language
my_dict = {'Python': 120, 'Java': 90, 'MySQL': 40, 'Php': 20}
prog_lang = list(my_dict.keys())
time_spend = list(my_dict.values())
```

```python
# Plot the bar chart
plt.bar(prog_lang, time_spend, color='grey', width=0.5)
plt.xlabel('Programming Languages')
plt.ylabel('Time Spent (hours)')
plt.title('Time Spent on Programming Languages')
plt.show()
```

```python
#MULTIPLE bar PLOT
py=(120,90,110,100,120)
ja=(80,90,75,85,95)
sql=(50,40,55,60,75)
php=(20,30,35,25,28)

xpos= np.arange(len(py))
xpos
```

```python
plt.figure(figsize=(10,7))
barwidth=0.2
plt.bar(xpos, py, color='red', width=barwidth,label='Python' )
plt.bar(xpos+0.2, ja, color='blue', width=barwidth,label='Java' )
plt.bar(xpos+0.4, sql, color='pink', width=barwidth,label='My SQL' )
plt.bar(xpos+0.6, php, color='yellow', width=barwidth,label='Php' )
plt.xticks(xpos+0.3,('Mon', 'Tues', 'Wed', 'Thurs', 'Fri'))
plt.xlabel('Programming Languages')
plt.ylabel('Time Spent (hours)')
plt.title('Time Spent on Programming Languages')
plt.show()
```

```python
#Stack Bar Plot
py=(120,90,110,100,120)
ja=(80,90,75,85,95)
php=(20,30,35,25,28)

x=['Mon', 'Tues', 'Wed', 'Thurs', 'Fri']
xpos=np.arange(len(x))
barwidth=0.4
plt.figure(figsize=(10,7))
plt.bar(xpos , py, color='pink', width=barwidth)
plt.bar(xpos , ja, bottom=py , color='purple', width=barwidth)
plt.bar(xpos , sql, bottom=ja , color='red', width=barwidth)


plt.xlabel('Programming Languages')
plt.ylabel('Time Spent (hours)')
plt.title('Time Spent on Programming Languages')
plt.xticks(xpos,x)
plt.show()
```
